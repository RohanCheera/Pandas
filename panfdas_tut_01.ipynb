From the above data we need to change the data types of the below 2 columns so our data will be handled easily.
* The unique values in WindGustSpeed column are : [30. 39. 85. 54. 50. 44. 43. 41. 48. 31. 46. 52. 33. 59. 26. 28. 35. 22.
 76. 65. 57. 69. 61. 24. 83. 63. 17. 20. nan 15. 13. 70. 78. 67. 98. 80.] and data type is float64

 --> convert into int64

* The unique values in WindSpeed9am column are : [ 6.  4. 30. 20. 19. 11.  7.  2. 13.  9. 15. 22. 41. 24. 17. nan  0. 26.
 33. 31. 28. 39. 35.] and data type is float64

 --> convert into int64


``Note:``The Null values are not converted to in to assigned data type directly,before converting the data type we need to fill null values/ replace null values with appropriate values.eg: mean, median, mode or any other values


```python
data['WindGustSpeed'].fillna(data['WindGustSpeed'].mean(),inplace=True)
data['WindSpeed9am'].fillna(data['WindSpeed9am'].mean(),inplace=True)
```

```python
data['WindGustSpeed']=data['WindGustSpeed'].astype('int64')
data['WindSpeed9am']=data['WindSpeed9am'].astype('int64')
```

```python
data.info()
```

```python

def get_cat_cols(data)->list:
  cat_cols=[]
  for i in data.columns:
    if data[i].dtype=='object':
      cat_cols.append(i)
  return cat_cols

cat_cols=get_cat_cols(data)
cat_cols
```

```python
for i in cat_cols:
  print(f"\nThe unique values in {i} column are :{data[i].unique()} \n")

  plt.figure(figsize=(8, 4))
  data[i].value_counts().plot(kind='bar', color='skyblue')

  # Add title and labels
  plt.title('Value Counts of Category Column')
  plt.xlabel(f"{i}")
  plt.ylabel('Count')
  plt.show()
  print("_"*100)
```
